```python
import requests

url =  'https://jsonplaceholder.typicode.com/todos/'
url2 = 'http://universities.hipolabs.com/search?country=United+States'

response = requests.get(url2)

if response.status_code == 200:
    data = response.json()
    print(data)
else:
    print('Failed in received data',response.status_code)
```

```python
import pandas as pd
df = pd.DataFrame(data)
```

```python
print(df)
```

```python
df.to_csv("US-data.csv")
```

```python
df.to_html("us-data.html")
```
