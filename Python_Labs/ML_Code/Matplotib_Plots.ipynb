# Basic Components or Parts of Matplotlib Figure


The parts of a Matplotlib figure include (as shown in the figure above):

Figure: The overarching container that holds all plot elements, acting as the canvas for visualizations.

Axes: The areas within the figure where data is plotted; each figure can contain multiple axes.

Axis: Represents the x-axis and y-axis, defining limits, tick locations, and labels for data interpretation.

Lines and Markers: Lines connect data points to show trends, while markers denote individual data points in plots like scatter plots.

Title and Labels: The title provides context for the plot, while axis labels describe what data is being represented on each axis.


1. Line Graph


```python
import matplotlib.pyplot as plt 
x = [3, 1, 3] 
y = [3, 2, 1] 

plt.plot(x, y)
plt.title("Line Chart")

plt.legend(["Line"])
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [0, 2, 4, 6, 8]
y = [0, 4, 16, 36, 64]

fig, ax = plt.subplots()  
ax.plot(x, y, marker='o', label="Data Points")

ax.set_title("Basic Components of Matplotlib Figure")
ax.set_xlabel("X-Axis") 
ax.set_ylabel("Y-Axis")  


plt.show()
```

2. Bar chart


```python
import matplotlib.pyplot as plt 
x = [3, 1, 3, 6, 2, 5, 4] 
y = [3, 2, 1, 4, 5, 6, 7] 

plt.bar(x, y)

plt.title("Bar Chart")

plt.legend(["bar"])
plt.show()
```

3. Plotting Histogram


```python
import matplotlib.pyplot as plt 
x = [1, 2, 3, 4, 5, 6, 7, 4] 

plt.hist(x, bins = [1, 2, 3, 4, 5, 6, 7])

plt.title("Histogram")
plt.legend(["bar"])

plt.show()
```

4. Scatter Plot


```python
import matplotlib.pyplot as plt 

x = [1,3,4,5,6,7,8]
y = [3, 2, 1, 4, 5, 6, 7]
plt.scatter(x, y)

plt.legend("A")

plt.title("Scatter chart")
plt.show()
```

6. Box Plot


```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(10)
data = np.random.normal(100, 20, 200)

fig = plt.figure(figsize =(10, 7))

plt.boxplot(data)

plt.show()
```

7. Stack Plot


```python
import matplotlib.pyplot as plt

days = [1, 2, 3, 4, 5]
Studying = [7, 8, 6, 11, 7]
playing = [8, 5, 7, 8, 13]

plt.stackplot(days, Studying, playing, colors =['r', 'c'])
plt.xlabel('Days')


plt.ylabel('No of Hours')

plt.title('Representation of Study and \
Playing wrt to Days')

plt.show()
```

```python
import pandas as pd

djia_data = pd.read_csv('./dataset/tesla-stock-price.csv')
djia_data.head()
```

```python
djia_data = djia_data.rename(columns = {' Open': 'Open', ' High': 'High', ' Low': 'Low', ' Close': 'Close'})
```

```python
djia_data['Date'] = pd.to_datetime(djia_data['Date'])
djia_data = djia_data.sort_values(by = 'Date')
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from datetime import datetime
```

Line Plots with a Single Line


```python
plt.plot(djia_data['Date'], djia_data['Close'])
plt.show()
```

Line Plots with Multiple Lines


```python
plt.plot(djia_data['Date'], djia_data['Open'])
plt.plot(djia_data['Date'], djia_data['Close'])


plt.show()
```

Adding a Legend


```python
plt.plot(djia_data['Date'], djia_data['Open'], label = 'Open')
plt.plot(djia_data['Date'], djia_data['Close'], label = 'Close')
plt.legend()
plt.show()
```

Drawing Bar Plots


```python
import pandas as pd
from calendar import month_name

# Convert Date column to datetime if not already
djia_data['Date'] = pd.to_datetime(djia_data['Date'])

# Extract and order months correctly
djia_data['Month'] = pd.Categorical(
    djia_data['Date'].dt.month_name(), 
    categories=list(month_name[1:]), 
    ordered=True
)

# Group metrics by monthly averages with observed=False
djia_monthly_mean = (
    djia_data.groupby('Month', observed=False)
    .mean()
    .reset_index()
)

djia_monthly_mean.head(6)
```

```python
plt.bar(djia_monthly_mean['Month'], height = djia_monthly_mean['Close'])
plt.show()
```
